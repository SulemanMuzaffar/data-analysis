```python
import pandas as pd
df = pd.read_csv('sales-data-sample.csv') #read csv file
df.head(5)
```

```python
columns = list(df.columns) #put columns into a new dataframe
columns
```

```python
df_datetime = pd.to_datetime(df["OrderDate"]) ##Cast to Date time format
df_datetime_o = pd.to_datetime(df["ShipDate"]) ##Case to date time format
```

```python
df_date = df_datetime.dt.date  ##Removing Time from Datetime dataframe
df_date.head()
df_date_o = df_datetime.dt.date
df_date_o.head()
```

```python
df['Order Date'] = df_date ##Appending the column to df_new
df['Shipping Date'] = df_date_o
```

```python
df = df.drop(['OrderDate', 'ShipDate', 'OrderProfitable', 'SalesaboveTarget'],axis="columns") ##Drop Extra Columns
```

```python
df = df.infer_objects() ##infer objects to specific datatypes
df.head()
```

```python
df.info() ##check for the datatypes of dataframe
```

```python
df.to_csv('CleanedSalesData.csv',index='false') ##Export the cleaned data to csv
```
